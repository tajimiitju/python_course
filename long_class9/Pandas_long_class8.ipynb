### Task : Which country has won the most men's gold medals in badminton over the years? Sort the results alphabetically by the player's  names.

```python
nn[(nn.Medal=='Gold')&(nn.Sport=='Badminton')&(nn.Gender=='Men')].sort_values('Athlete')
```

### Task : Display the male gold medal winners for the 100m Track event over the years. List the results starting with the most recent. Show the Olympic City, Edition, Athlete and the country they represent.

```python
nn[(nn.Medal=='Gold')&(nn.Event.str.contains('100m'))&(nn.Gender=='Men')].head(5)
```

```python
nn[(nn.Medal=='Gold')&(nn.Event.str.contains('100m'))&(nn.Gender=='Men')].sort_values('Edition',ascending=False).head(10)
```

```python
nn[(nn.Medal=='Gold')&(nn.Event.str.contains('100m'))&(nn.Gender=='Men')].sort_values('Edition',ascending=False)[['City', 'Edition', 'Athlete','NOC']]
```
