## Maps

map takes every value in the column it is being called on and converts it some new value using a function you provide it.

A "map" is a term, borrowed from mathematics, for a function that takes one set of values and "maps" them to another set of values. In data science we often have a need for creating new representations from existing data, or for transforming data from the format it is in now to the format that we want it to be in later. Maps are what handle this work, making them extremely important for getting your work done!

There are two mapping functions that you will use often. The Series map is the first, and slightly simpler one. For example, suppose that we wanted to remean the scores the wines recieved to 0.

```python
mean_points = pp['points'].mean()
mean_points
#map(function, variables list)
```

```python
pp['points'].map(lambda x: x-mean_points) #ekihane lambda holo function r variable list holo pp['points']
```

## Apply
map takes a Series as input. The DataFrame apply function can be used to do the same thing across columns, on the level of the entire dataset.

```python
#map er motoi kaj kore
#map only amake series and work on series
#apply can work on data frame and series both
pp['points'].apply(lambda x: x-mean_points)
```

```python
def dif(data_frame):
    data_frame['points'] = data_frame['points'] - mean_points
    return data_frame
pp.apply(dif,axis='columns')
```

```python
#pp.map(dif,axis='columns') #kaj korbena... tai amra always apply use korbo
```

```python
# country r province mile akta series bana
pp[['country','province']].apply(lambda x: x['country']+x['province'],axis='columns')
```

#### Remap the `price` column by subtracting the median price. Use the `Series.map` method

```python
#return a series
median_price = pp['price'].median()
median_price
pp['price'].map(lambda x: x-median_price)
```

#### Remap the `price` column by subtracting the median price. Use the `DataFrame.apply` method.

```python
#return a series
def dif(data_frame):
    data_frame['price'] = data_frame['price'] - median_price
    return data_frame['price'] # return data_frame dile data frame return korto
pp.apply(dif,axis='columns')
```
